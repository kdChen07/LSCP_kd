```python
from access import Access, weights, datasets
import logging
import pandas as pd
import geopandas as gpd
from shapely import wkt
import requests
import json
from google.colab import files
```

```python
MSOA=pd.read_csv('https://github.com/kdChen07/progress-and-problems/blob/main/data/Point%20MSOAs%20Lack%20of%20Vaccination%20Service.csv?raw=true')
MSOA.head(1)
```

```python
MSOA['geometry'] = MSOA['geometry'].apply(wkt.loads)
MSOA_lack = gpd.GeoDataFrame(MSOA, geometry='geometry')
MSOA_lack.crs = 'epsg:27700'
```

```python
MSOA_lack = MSOA_lack.to_crs('epsg:4326')
MSOA_lack
```

```python
MSOA_lack['MSOA_lon'] = MSOA_lack['geometry'].x
MSOA_lack['MSOA_lat'] = MSOA_lack['geometry'].y
MSOA_lack['demand_df'] = 1
```

```python
MSOA_lack
```

```python
MSOA_lack_df = MSOA_lack[['msoa11cd','MSOA_lon','MSOA_lat']]
MSOA_lack_df
```

```python
df=pd.read_csv('https://github.com/kdChen07/progress-and-problems/blob/main/data/poi_three_types1.csv?raw=true')
df.head(1)
```

```python
point = gpd.GeoDataFrame(df, geometry=gpd.points_from_xy(df.feature_easting, df.feature_northing))
point.crs = 'epsg:27700'
```

```python
point
```

```python
point = point.to_crs('epsg:4326')
point['point_lon'] = point['geometry'].x
point['point_lat'] = point['geometry'].y
```

```python
point['supply_value']=1
```

```python
point
```

```python
point_df = point[['index','point_lon','point_lat']]
point_df
```

```python
distance = Access(demand_df = MSOA_lack,
           demand_index='msoa11cd',
           demand_value='demand_df',
           supply_df= point,
           supply_index= 'index',
           supply_value=['supply_value']
                 )

distance.demand_df = distance.demand_df.to_crs('epsg:27700')
distance.supply_df = distance.supply_df.to_crs('epsg:27700')
```

```python
distance.create_euclidean_distance(threshold = 16093.44)
distance.cost_df
```

```python
euclidean_distance_matrix = distance.cost_df
```

```python

euclidean_distance_matrix = pd.merge(left = euclidean_distance_matrix, 
                                     right = point_df, 
                                     how = 'left', 
                                     left_on = 'dest', 
                                     right_on = 'index')
euclidean_distance_matrix = pd.merge(left = euclidean_distance_matrix, 
                                     right = MSOA_lack_df, 
                                     how = 'left', 
                                     left_on = 'origin', 
                                     right_on = 'msoa11cd')
```

```python
euclidean_distance_matrix
```

```python
#euclidean_distance_matrix.drop(columns=['msoa11cd','name'], axis=1,inplace=True)
#euclidean_distance_matrix
```

```python
euclidean_distance_matrix.shape
```

```python
euclidean_distance_matrix_sub01=euclidean_distance_matrix[0:3]
```

```python
euclidean_distance_matrix_sub01
```

```python
distance_list = []
origin_id_list = []
destination_id_list = []

for (i1, row1) in euclidean_distance_matrix_sub01.iterrows():
    LatOrigin = str(row1['MSOA_lat'])
    LongOrigin = str(row1['MSOA_lon'])
    origin_id = row1['origin']
    
    LatDestination = str(row1['point_lat'])
    LongDestination = str(row1['point_lon'])
    destination_id = row1['dest']
    
    url = "http://82.11.51.26:5000/route/v1/driving/" + LongOrigin +","+ LatOrigin + ";" + LongDestination +","+ LatDestination + "?steps=false"
    result_distance = requests.get(url).json()['routes'][0]['distance']
    distance_list.append(result_distance)
    origin_id_list.append(origin_id)
    destination_id_list.append(destination_id)

output01 = pd.DataFrame(distance_list, columns = ['Distance in meter'])
output01['origin_id'] = origin_id_list
output01['destination_id'] = destination_id_list
output01
```

```python
output01.to_csv('distance18994.csv')
```
